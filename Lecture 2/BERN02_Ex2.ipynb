## Task 1
The above table cannot be readily processed for various reasons. 

- Firstly, the numbers are actually strings and contain the unicode character ±.
- Secondly, values and errors are mixed. 


```python
type(df['Site'])
```

```python
df.pKa.str.split("±").str.get(0)
```

```python
df[['Site']]
```

```python
df = df.rename(columns={'Î´low': 'low',
                   'Î´high': 'high'})
df
```

```python
type(df['pKa'][0])
```

```python
df[['pKa', 'err']] = df['pKa'].str.split(r'\s*Â±\s*',expand=True).astype(float)
```

```python
df['pKa'][0]
```

```python
df
```

```python
df[['resname', 'resnum']] = df['Site'].str.extract(r'([A-Za-z]+)(\d+)')
```

```python
df = df[['pKa', 'err', 'resname', 'resnum']]
df
```

```python
type(df['resnum'][0])
```

```python
df.loc[:,'resnum'] = pd.to_numeric(df['resnum'], errors='coerce').astype("Int64")
df
```

```python
df.to_csv("croke.csv", index=False)
```
