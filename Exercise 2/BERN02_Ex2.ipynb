## Discussion
### How the original paper relates to the FAIR data principles:
#### Finable
- The article is published in Protein Science, a peer-reviewed journal, with a DOI, which makes the publication itself finable
- However the datasets and plots used in the article are only presented in tables or pdf format and are not deposited in public repositories like Zenodo. This makes the raw and processed data less finable for reanalysis.

#### Accessible
- The paper is Free Access, so the text, tables, and figures are available to all readers without a paywall barrier.
- But the raw and processed experimental data are not openly available in dedicated repository, which limits the accessibility of the data.

#### Interopable
- The results are reported in tables and plots but not in machine-readdable formats. This is great for human but not usable for machines.

#### Reusable
- The scientific results are resuable for comparison and citation.
- However, without raw datasets, metadata, or standardized file formats, the degree of reusability is limited. Future researchers cannot refit or reproduce the analysis from the original measurement.



### Detail how my workflow is an improvement with regard to the FAIR principles.
#### Findable
- In my workflow, I import the paper as HTML, extract the data, and save it in a structured CSV format. This means that the data are now stored in a machine-readable file with clear column names, so they can be indexed, searched, and found more easily by both humans and machines.

#### Accessible
- My cleaned CSV file is shared openly on GitHub respository. That ensures the data are not only visible but also directly downloadable and resuable.

#### Interopable
- By providing documentation (metadata) alongside the csv, others can replicate and extend the research.

#### Reusable

## Task 3

The acid dissocitation constant of an isolated glutamic acid ("Glu") is p =4.1. As a function of residue number, plot the shift from this reference value for all glutamates in the above tabel. Hints: Use matplotlib and a pandas mask using e.g. mask = df.resname == "Glu".

```python
df[df['resname'] == 'Glu']
```

```python
glu_mask =df['resname'] == 'Glu'
glu_df = df[glu_mask]
glu_df
```

```python
pka_ref = 4.1
glu_df['shift'] = glu_df['pKa'] - pka_ref
glu_df
```

```python
plt.figure(figsize=(10,6))
plt.errorbar(glu_df['resnum'], glu_df['shift'], yerr=glu_df['err'], fmt='o', capsize=5)
plt.axhline(0, color="gray", linestyle="--", label="Reference pKa = 4.1")
plt.xlabel('Residue Number')
plt.ylabel('pKa Shift (from 4.1)')
plt.title('pKa Shifts from the acid dissocitation constant of an isolated glutamic acid (pKa = 4.1)')
plt.legend()
plt.show()
```
