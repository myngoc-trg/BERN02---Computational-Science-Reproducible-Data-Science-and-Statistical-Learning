```python
import pandas as pd

url = "https://raw.githubusercontent.com/myngoc-trg/STAN48-Programming-for-Data-Science/refs/heads/main/Final%20Project/Data/cleaned_credit_risk_dataset.csv"


df = pd.read_csv(url,index_col=0)

# Display the first few rows
df.head()
```

```python
#VIF

X = df

X = sm.add_constant(X)

#Compute VIF

vif_data = pd.DataFrame()

vif_data["feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data
```

```python
vif_filteredData = vif_data[vif_data['VIF']<10]

nummeric_collumns = vif_filteredData['feature'].astype(str).tolist()


filtered_data = df[nummeric_collumns]

filtered_data.head()
```

```python
corr_matrix = filtered_data.corr()

plt.figure(figsize=(8, 6))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths = 0.5, vmin = -1, vmax =1)

plt.title('Correlation Matrix')
```
