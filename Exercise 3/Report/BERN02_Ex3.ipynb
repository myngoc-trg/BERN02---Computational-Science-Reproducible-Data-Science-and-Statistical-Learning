```python
class LocalRegression():
    def __init__(self, X, Y):
        self.X = np.array(X)
        self.Y = np.array(Y)
    def local_regression(self,x0, k):
        """
        X: vector of obs of predictor
        Y: vector of obs of response
        x0: vector values for which a prediction is going to made
        k: number of nearest neighbors    
        """
        # add first row 1 for the intercept theta_0
        #x0 = np.array(x0)
        if isinstance(x0, (int, float)):
            x0 = np.array([x0])  # convert to 1D array with 1 element

        
        # Solving the OLS
        #print(X_weights)
        x0 = np.r_[1, x0]
        #x0 = [1,x0]
        X = np.c_[np.ones(len(self.X)), self.X]
        k_X = X[:k]
        k_Y = self.Y[:k]
        X_weights = k_X.T @ self.weight_func(X,x0, k=k)
        thetas = np.linalg.pinv(X_weights @ k_X) @ X_weights @ k_Y
        #print("X = ", X)
        print("Estimated parameters: ",thetas)
        pred = x0 @ thetas
        residuals = self.Y - X @ thetas 
        sigma_2_est = (residuals @ residuals) / (X.shape[0] - X.shape[1])
        se = np.sqrt(sigma_2_est * (x0 @ np.linalg.pinv(X_weights @ k_X) @ x0))
        return pred, se
    def weight_func(self, X,x0, k):
        weights = []
        distances = []
        for row in X:
            eucl_distance = np.linalg.norm(row - x0)
            distances.append(1/eucl_distance)
        distances = sorted(distances)[:k]
        distances_inv = [1/d for d in distances]
        return np.diag(distances_inv)
```
